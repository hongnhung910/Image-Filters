OpenCV provides four variations of this technique.

1. __cv2.fastNlMeansDenoising(src, dst, h, hColor, templateWindowSize, searchWindowSize)__ : works with a single grayscale images
    
2. __cv2.fastNlMeansDenoisingColored(src, dst, h, hColor, templateWindowSize, searchWindowSize)__ : works with a color image.
    
3. __cv2.fastNlMeansDenoisingMulti(src, dst, h, hColor, templateWindowSize, searchWindowSize)__ : works with image sequence captured in short period of time (grayscale images)
    
4. __cv2.fastNlMeansDenoisingColoredMulti(src, dst, h, hColor, templateWindowSize, searchWindowSize)__ : same as above, but for color images.

Common arguments are:
 ```markdown  
* src : Input 8-bit 3-channel image.
* dst : Output image with the same size and type as src .    
* h : Parameter regulating filter strength for luminance component. Bigger h value perfectly removes noise but also removes image details, smaller h value preserves details but also preserves some noise.
* hColor : same as h, but for color images only. (normally same as h)
* templateWindowSize : Size in pixels of the template patch that is used to compute weights. Should be odd. Recommended value 7 pixels
* searchWindowSize : Size in pixels of the window that is used to compute weighted average for given pixel. Should be odd. Affect performance linearly: greater searchWindowsSize - greater denoising time. Recommended value 21 pixels
    
  ```

````python
import numpy as np
import cv2
from matplotlib import pyplot as plt
````

````python
# Denoise color image
img = cv2.imread("./Data/NoiseImage.jpg")
dst = cv2.fastNlMeansDenoisingColored(img,None,40,40,7,21)

plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(dst)
plt.show()
````

````python
# Denoise grayscale image
img = cv2.imread("./Data/city.jpg")
dst = cv2.fastNlMeansDenoising(img,None,20,7,21)

plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(dst)
plt.show()
````
